```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Feature Engineering -> working on 'Date_of_Journey' column
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
# Drop 'Date_of_Journey' column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
# retrieving hours and minutes from 'Arrival_Time'
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split()[0])
```

```python
df.head()
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1]

df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
```

```python
df.head()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
df.head()
```

```python
# working on 'Dep_Time'
df['Departure_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_Min'] = df['Dep_Time'].str.split(':').str[1]

df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Min'] = df['Departure_Min'].astype(int)

df.head()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Working on total stops
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
# Let's replace nan with 1
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1
})

df[df['Total_Stops'].isnull()] # No nan values
```

```python
# drop route(We already have source and destination)
df.drop('Route',axis=1,inplace=True)
df.head()
```

```python
# Working on 'Duration'
df['Duration_hours'] = df['Duration'].str.split().str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split().str[1].str.split('m').str[0]
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
df.head(2)
```

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
df['Additional_Info'].unique()
```

```python
df.drop('Additional_Info',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
temp_df = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
# merge temp_df with original df and drop 'Airline','Source','Destination'
# so that the data will be ready to be fed to our model
```

```python
final_df = pd.concat([df,temp_df],axis=1) # cancat side by side
```

```python
final_df.head()
```

```python
final_df.drop('Airline',axis=1,inplace=True)
final_df.drop('Source',axis=1,inplace=True)
final_df.drop('Destination',axis=1,inplace=True)
```
